```python
# Deze cell importeert een aantal functies en modules die je nodig hebt om deze opdracht te maken.

from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

# Start notebook mode
init_notebook_mode(connected=True)

import pandas as pd # we gebruiken pandas om de CSV data te laden en te visualiseren
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding = "ISO-8859-1")
df.head(n=10)
```

```python
df_aantal_per_jaar = df['iyear'].value_counts().sort_index()
```

```python
trace = go.Scatter(
    x=df_aantal_per_jaar.index,
    y=df_aantal_per_jaar.values
)
iplot([trace])
```

```python
df_aantal_succes = df[df['success'] == 1]
df_aantal_succes_pj = df_aantal_succes['iyear'].value_counts().sort_index()
```

```python
trace = go.Scatter(
    x=df_aantal_per_jaar.index,
    y=df_aantal_per_jaar.values,
    name='Totaal aantal aanslagen'
)
trace2 = go.Scatter(
    x=df_aantal_succes_pj.index,
    y=df_aantal_succes_pj.values,
    name='Succesvolle aanslagen'
)
iplot([trace,trace2])
```

```python
df_region = df['region_txt'].value_counts()
```

```python
trace = go.Bar(
    x=df_region.index,
    y=df_region.values
)
iplot([trace])
```

```python
df_w_eu = df[df['region_txt'] == 'Western Europe']
df_aantal_per_jaar_eu = df_w_eu['iyear'].value_counts().sort_index()

df_aantal_succes_eu = df_w_eu[df_w_eu['success'] == 1]
df_aantal_succes_pj_eu = df_aantal_succes_eu['iyear'].value_counts().sort_index()
```

```python
trace = go.Scatter(
    x=df_aantal_per_jaar_eu.index,
    y=df_aantal_per_jaar_eu.values,
    name='Totaal aantal aanslagen'
)
trace2 = go.Scatter(
    x=df_aantal_succes_pj_eu.index,
    y=df_aantal_succes_pj_eu.values,
    name='Succesvolle aanslagen'
)
iplot([trace,trace2])
```

```python
from numpy import arange,array,ones
from scipy import stats

xi = df_aantal_succes_pj_eu.index
y = df_aantal_succes_pj_eu.values

slope, intercept, r_value, p_value, std_err = stats.linregress(xi,y)
line = slope*xi+intercept


trace1 = go.Scatter(
                  x=xi,
                  y=y,
                  mode='markers',
                  name='Data'
                  )
trace2 = go.Scatter(
                  x=xi,
                  y=line,
                  mode='lines',
                  name='Fit'
                  )
iplot([trace1,trace2])
```

```python
df_eu_grtr_2010 = df_aantal_succes_pj_eu[df_aantal_succes_pj_eu.index >= 2010]


xi = df_eu_grtr_2010.index
y = df_eu_grtr_2010.values

slope, intercept, r_value, p_value, std_err = stats.linregress(xi,y)
line = slope*xi+intercept


trace1 = go.Scatter(
                  x=xi,
                  y=y,
                  mode='markers',
                  name='Data'
                  )
trace2 = go.Scatter(
                  x=xi,
                  y=line,
                  mode='lines',
                  name='Fit'
                  )
iplot([trace1,trace2])
```
